```python
#STRING

name = "John"
age = 19
print(name, type(name))
print(age, type(age))

print("My name is " + name + " and I am " + str(age) + " years old")

print("My name is {name} and I am {age} years old".format(name=name, age=age))

print(f"My name is {name} and I am {age} years old")

s = 'Amazing news'

#1. len() - length of string
print(len(s))

#2. capitalize() - 1st letter capital
d = 'i love india'
print(d.capitalize())

#3. title() - All 1st letters of words = capital
print(d.title())

#4. find(substr,start,end) - Finds 1st index where the given letter/str is present, -1 if absent
print(d.find('a'))
print(d.find('z'))

#5. index(substr,start,end) - Finds 1st index where the given letter/str is present, error if absent
print(d.find('l'))

#6. isalnum() - True if string has alphabets or numbers
print(d.isalnum())
print(s.isalnum())
print()

#7. isalpha() - True if string has only alpha
print(d.isalpha())
print(s.isalpha())

#8. isdigit() - True if string has only nums
print(d.isdigit())
print(s.isdigit())

#9. isupper() - Checks if all alphabets = Uppercase
print('HEL@#LO123'.isupper())

#10. islower() - Checks if all alphabets = Lowercase
print('hello1234567@'.islower())

#11. lower() - converts to lowercase
print(s.lower())

#12. upper() - Converts to uppercase
print(d.upper())

#13. startswith(substr,start,end) - Checks if given string starts with given substring
print(d.startswith('a'))

#14. endswith(substr,start,end) - Checks if given string ends with given substring
print(d.endswith('a'))

#15. replace() -  Replaces the substring with new one
print(d.replace(' ','12'))

#16. join() - Joins external string to given
print('*'.join(d))
print('*'.join(['hello',' ' ,'bye']))

#17. split() - Splits the given string at given substring into a list
print(d.split(' '))

#18. partition() - Splits the string at given substring ONLY AT FIRST OCCURENCE, len of list is always = 3
print(d.partition(' '))

#19. swapcase() - Swaps upper to lower and lower to upper case
print(d.title().swapcase())

#20. count(substr,start,end) - Counts substr
print(d.count(' '))

#Reverse a string
print(d[::-1])
```

```python
#SETS

s = {12,23,34,45,56,57}
print(s, type(s), id(s))

print(*s)

s.add("Hi")
print(s)

s.remove(23)
print(s)

print(id(s))

s = {1,2,3,4,5}
s2 = {1,2,3}


#1. .difference(set2) - Gives difference between both sets i.e elements in set1 but not in set2
print(s.difference(s2))
print(s2.difference(s))

#OR
print(s - s2)
print(s2 - s)


#2. .intersection(set2) - Gives commons elements between set1 and set2
print(s.intersection(s2))
print(s2.intersection(s))

#OR
print(s & s2)
print(s2 & s)


#3. .union(set2) - Gives a set with elements of set1 and set2
s3 = {12,23,34}
print(s3, s)

print(s.union(s3))

#OR

print(s |s3)

print(s)
```

```python
#DICTIONARIES

d = {"username":"PizZaBOi", "email":"abc.xyz@gmail.com", "pfp":False}
print(d, type(d), id(d))

d2 = dict(name="PizZaBOi", altname=False)
print(d2, type(d2), id(d2))

print(d)

d["age"] = 19
print(d)

print(d["username"])

d3 = dict(zip(["a","b",'c'],[12,23,34]))
print(d3, type(d3), id(d3))

#1. len() - Length
print(len(d))

#2. get() - gets values of a key
print(d.get('email'))
    #EQUIVALENT TO
print(d['email'])

#3. items() - gives all the key,value pairs in their independent tuples
print(d.items())
for x in d.items():
    print(x)

#4. keys() - gives list of all keys
print(d.keys())
for x in d.keys():
    print(x)

#5. values() = gives list of all values
print(d.values())
for x in d.values():
    print(x)

#6. dict.fromkeys(x,y) - Assigns common value 'y' to all the values of the sequence 'x'
n = d.fromkeys([12,23,34],'common value')
print(n)

n2 = dict.fromkeys([112,213,314],'common value')
print(n2)

#7. setdefault() - Adds new key:value pair to the ditionary if key is unique, else no change
d.setdefault('table',3)
print(d)

#8. update() - Merges key:value pair from new dictionary into original
d.update(n)
print(d, n, sep="\n\n")

#9. pop() - Removes key:value pair, given input in key along with a custom message if key is absent
print(d.pop('table', 'key not found for table'))
print(d)

print(d.pop(12, 'key not found for 12'))
print(d)

print(d.pop('new table', 'key not found for new table'))
print(d)

#10. popitem() - Removes last entered key:value pair in form of tuple, error if dict in empty
print(d.popitem())
print(d)
```

```python
#CLASSES OBJECTS AND FILES

class fileMani:
    def __init__(self, filename, text):
        self.filename = filename
        self.text = self.fileContent = text

        self.f = open(f"{filename}.txt", "w")
        self.f.write(self.text)
        self.f.close()
        
    def writeFile(self, text):
        self.f = open(f"{self.filename}.txt", "a+")
        self.f.write(text)
        self.f.close()

    def readFile(self):
        self.f = open(f"{self.filename}.txt", "r")
        self.fileContent = self.f.read()
        self.f.close()
        return self.fileContent

    def eraseFile(self):
        self.f = open(f"{self.filename}.txt", "w")
        self.text = ''
        self.f.write(self.text)
        self.f.close()
    
    def readFileLines(self):
        self.f = open(f"{self.filename}.txt", "r")
        self.fileLines = self.f.readlines()
        self.f.close()
        return self.fileLines
    
    def fileReplace(self, old, new):
        self.readFile()
        self.fileContent = self.fileContent.replace(old, new)
        self.eraseFile()
        self.writeFile(self.fileContent)

    def fileDelLine(self, line):
        self.readFileLines()
        self.fileLines.pop(line-1)
        self.eraseFile()
        self.text = ''.join(LINE for LINE in self.fileLines)
        self.writeFile(self.text)

    def fileInsertLine(self, line, lineText):
        self.readFileLines()
        self.fileLines.insert(line-1, lineText+"\n")
        self.eraseFile()
        self.text = ''.join(LINE for LINE in self.fileLines)
        self.writeFile(self.text)

fileObj = fileMani(text = "Hello World", filename = "newFile")
fileObj.writeFile("\nHello")
fileObj.writeFile("\nApple")
fileObj.writeFile("\nBanana")
fileObj.writeFile("\nSpaghetti")

fileObj.fileReplace("Apple", "Mango")
fileObj.fileInsertLine(2, "LOL LOL LOL")
fileObj.fileDelLine(1)
fileObj.fileInsertLine(3, "\tMeow")
```
